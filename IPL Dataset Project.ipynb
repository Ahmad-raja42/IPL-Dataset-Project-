```python
#Loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading the ipl matches dataset
ipl=pd.read_csv('matches.csv')
```

```python
ipl.head()
```

```python
# The number of rows and columns in the dataset
ipl.shape
```

```python
ipl['player_of_match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
# A bar-plot for the top 5 players with most man of the match awards
plt.figure(figsize=(5,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color=["blue","yellow","orange","green","red"])
plt.show()
```

```python
# The frequency of results
ipl['result'].value_counts()
```

```python
#Finding out the number of toss wins each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python
#Making a histogram 
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
#Finding out the number of wins each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 5 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=["blue","yellow","orange","green","red"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# Extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=20)
plt.show()
```

```python
#Finding out the frequency of number of wins each time after batting second
batting_second['winner'].value_counts()
```

```python
#Making a bar plot for top-5 teams with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]),color=["red","yellow","blue","green","orange"])
plt.show()
```

```python
# Making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
